## Example -- Plotting

In this example we show how to plot data

```julia
using Plots
```

## Example 1 

In this example, we will show how to plot $\sin(x)$.
* First, we want to plot $\sin(x)$, so we input sin into the first block.
* Second, we set the range from $-2\pi$ to $\pi$. Third, we label the image.

```julia
plot(sin, -2pi, pi, label="sine function")
```

## Example 2

In this example, we will show how to plot three-dimensional graphics.
* First, we set N=200 and d = 8*pi/N is a constant and n=[0,1,2,,,,,,,,200] is an 201 * 1 array.
* Second, x is $dn \times \cos(\theta)$, and the range of dn is from 0 to $8\pi$, the definition method of y is similar.
* Third, z= 0.0:N means that the range of z is from 0 to 200.
* Last, we use marker to set the style of the points in the image, and legend=false represent we dont mark grapic.

According to the definition of x,y,z we can clearly know that it is a spiral, the range of radius is from 0 to $8\pi$, and the range of angle is also from 0 to $8\pi$.

```julia
N = 200; d = 8*pi/N;
n= 0.0:N;
xx = d*n.*cos.(d*n);
x= [d*n*cos.(d*n) for n in 0:N]; 
y= [d*n*sin.(d*n) for n in 0:N];
z= 0.0:N;
plot( xx, y, z, marker=:circle, legend=false)
```

## Example 3

In this example, we will show how to add a graphic to an already drawn graphic.
* First, we set x to be intergers from 1 to 10, and y to be ten random numbers between 0 and 1, then we plot the image.
* Second, we set z to be random numbers between 0 and 1.
* Last, we overlap the second function graphic with the first graphic.

```julia
x = 1:10; y = rand(10);
plot(x,y)
```

```julia
z = rand(10)
plot!(x,z)
```

## Example 4

In this example, we will show how to plot two graphic at the same time.
* First, we set x to be the intergers from 1 to 10, and y to be a 10*2 matrix, and first column represents first function value, second column represents second function value.
* Second, we plot the graphic,and set the title and label function's name, lw=3 represents the Width of line is three.
* Third, xlabel!("My x label") means that the label of x-axis is "My x label"

```julia
x = 1:10; y = rand(10,2);
plot(x,y,title="Two Lines",label=["Line 1" "Line 2"],lw=3)
xlabel!("My x label")
```

## Example 5

In this example, we will show how to plot three function graph.
* First, we set x to be the intergers from 1 to 10, and y to be a 10*2 matrix, and first column represents first function value, second column represents second function value.
* Second, we plot the graphic.
* Third, we set z to be 10 random numbers between 0 and 1.
* Last, we overlap the second function graphic with the first graphic.

```julia
x = 1:10; y = rand(10,2);
p = plot(x,y)
```

```julia
z = rand(10)
plot!(p,x,z)
```
